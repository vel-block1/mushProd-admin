```python
import numpy as np
import pandas as pd
from sklearn.gaussian_process import GaussianProcessRegressor
from sklearn.gaussian_process.kernels import RBF
from sklearn.metrics import r2_score, mean_squared_error
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
import matplotlib.pyplot as plt
from sklearn.metrics import mean_absolute_error, r2_score
import joblib

# Read the data from CSV file
data = pd.read_csv('/Users/knt/Documents/Mush_Data_3.csv')

# Extract the features (temperature and humidity) and target variable (production rate)
X = data[['Number of bags','Temperature','Humidity']].values
y = data['Grams Harvested'].values

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4, random_state=42)

# Scale the data
scaler_X = StandardScaler()
scaler_y = StandardScaler()
X_train_scaled = scaler_X.fit_transform(X_train)
y_train_scaled = scaler_y.fit_transform(y_train.reshape(-1, 1))
X_test_scaled = scaler_X.transform(X_test)

# Define the Gaussian Process Regression model
kernel = RBF()
# Adjust the alpha parameter for regularization
gp_model = GaussianProcessRegressor(kernel=kernel, optimizer='fmin_l_bfgs_b', n_restarts_optimizer=10, alpha=0.01, normalize_y=True)

# Fit the model to the scaled training data
gp_model.fit(X_train_scaled, y_train_scaled)

# Make predictions on the training and testing data
y_train_pred_scaled = gp_model.predict(X_train_scaled)
y_train_pred = scaler_y.inverse_transform(y_train_pred_scaled.reshape(-1, 1)).flatten()

y_test_pred_scaled = gp_model.predict(X_test_scaled)
y_test_pred = scaler_y.inverse_transform(y_test_pred_scaled.reshape(-1, 1)).flatten()

# Calculate the R-squared value and mean squared error for training and testing data
r2_train = r2_score(y_train, y_train_pred)
mse_train = mean_squared_error(y_train, y_train_pred)

r2_test = r2_score(y_test, y_test_pred)
mse_test = mean_squared_error(y_test, y_test_pred)

print("Training Set - R-squared:", r2_train)
print("Training Set - Mean Squared Error:", mse_train)
print("Testing Set - R-squared:", r2_test)
print("Testing Set - Mean Squared Error:", mse_test)

# Data for R-squared and MSE
r2_train_test = [(r2_train, 'Training'), (r2_test, 'Testing')]
mse_train_test = [(mse_train, 'Training'), (mse_test, 'Testing')]

# Plotting
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 6))

ax1.set_title('R-squared')
ax1.set_xlabel('R-squared')
ax1.set_ylabel('Set')
ax1.barh(1, r2_train)
ax1.barh(2, r2_test)

ax2.set_title('Mean Squared Error')
ax2.set_xlabel('Mean Squared Error')
ax2.set_ylabel('Set')
ax2.barh(1, mse_train)
ax2.barh(2, mse_test)

plt.tight_layout()
plt.show()

# Calculate the MAE for training and testing data
mae_train = mean_absolute_error(y_train, y_train_pred)
mae_test = mean_absolute_error(y_test, y_test_pred)

# Calculate the R-squared value for training and testing data
r2_train = r2_score(y_train, y_train_pred)
r2_test = r2_score(y_test, y_test_pred)

print("Training Set - MAE:", mae_train)
print("Training Set - R-squared:", r2_train)
print("Testing Set - MAE:", mae_test)
print("Testing Set - R-squared:", r2_test)

# Data for MAE and R-squared
mae_train_test = [(mae_train, 'Training'), (mae_test, 'Testing')]
r2_train_test = [(r2_train, 'Training'), (r2_test, 'Testing')]

# Plotting
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 6))

ax1.set_title('Mean Absolute Error')
ax1.set_xlabel('Mean Absolute Error')
ax1.set_ylabel('Set')
ax1.barh(1, mae_train)
ax1.barh(2, mae_test)

ax2.set_title('R-squared')
ax2.set_xlabel('R-squared')
ax2.set_ylabel('Set')
ax2.barh(1, r2_train)
ax2.barh(2, r2_test)

plt.tight_layout()
plt.show()

# Scale the input data for prediction
input_data_scaled = scaler_X.transform([[120, 26.2, 96.3]])

# Make the prediction
predicted_weight_scaled = gp_model.predict(input_data_scaled)

# Inverse transform the predicted value to the original scale
predicted_weight = scaler_y.inverse_transform(predicted_weight_scaled.reshape(-1, 1)).flatten()[0]

print("Predicted Weight:", predicted_weight, "g")
```
